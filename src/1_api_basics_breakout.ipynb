```python
### Investigate the action space and observation space ###

env_name = 'ALE/Breakout-v5'
env = gym.make(env_name, render_mode='rgb_array')

observation, info = env.reset()

# Print info, obs space shape, number of possible actions, etc.
print(f'{info=}')
print(f'{observation.shape=}')
print('----------------------')
print(f'Observation space shape: {env.observation_space.shape=}')
print(f'{env.observation_space=}')
print('----------------------')
print(f'Number of possible actions: {env.action_space.n=}')
print(f'{env.action_space=}')

env.close()
```

```python
### Print screen at start (prior to any actions) ###

print(f'{info=}')

%matplotlib inline
plt.imshow(observation)
```

```python
### Perform a series of random actions (with display window and time delay. Note: render_mode='human' may cause issues in Jupyter (vs. a Python script)) ###

env_name = 'ALE/Breakout-v5'
env = gym.make(env_name, render_mode='human')

observation, info = env.reset()
for step in range(250):
    env.render()
    random_action = env.action_space.sample()
    observation, reward, done, truncated, info = env.step(random_action)
    print(f'{step=} | {random_action=} | {reward=} | {done=} | {truncated=} | {info=}')
    time.sleep(0.02) # Slow down steps to observe the rendering

    if done or truncated:
        break
        # observation, info = env.reset() # Could reset instead of breaking loop

env.close()
```

```python
### Perform a series of random actions (no display window or time delay) ###

env_name = 'ALE/Breakout-v5'
env = gym.make(env_name, render_mode='rgb_array')

observation, info = env.reset()
for step in range(250):
    random_action = env.action_space.sample()
    observation, reward, done, truncated, info = env.step(random_action)
    print(f'{step=} | {random_action=} | {reward=} | {done=} | {truncated=} | {info=}')

    if done or truncated:
        break
        # observation, info = env.reset() # Could reset instead of breaking loop

env.close()
```
